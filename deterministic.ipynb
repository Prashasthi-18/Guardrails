```python
def deterministic_approach(text: str)->bool:
    """Returns true if the content is blocked"""

    blocked_words = ["hack", "malware", "virus", "kill", "poison", "bomb", "hate", "exploit"]

    for word in blocked_words:
        if word in text.lower():
            return "Blocked"
    return "Allowed"
```

```python
test_input = [
    "How do i poison my neighbour?",
    "I want to hack my friends phone",
    "What is bomb calorimeter"
]

for input in test_input:
    print(input , ":", deterministic_approach(input))
```
