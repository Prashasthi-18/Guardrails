```python
# PII (Personally Identifiable Information) is any information that can identify a person.

# Examples include:

# Name
# Email address
# Phone number
# Credit card number
# Aadhaar number
# Passport number
# API keys
# Social Security Number (SSN)

# When building AI agents, we usually don't want sensitive information reaching the LLM unnecessarily, especially if the model doesn't need it.

# That's where PIIMiddleware comes in.
```

```python
import re
```

```python
def pii_middleware(user_input):

    # Email regex
    email_pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"

    # Replace email with **** 
    cleaned_text = re.sub(email_pattern, "********", user_input)

    return cleaned_text
```

```python
# Test
text = "My email is prashasthiboin@gmail.com. Please summarize my resume."

print("Original :", text)
print("Modified :", pii_middleware(text))
```

PHONE NUMBER + EMAIL

```python
import re

def pii_middleware(user_input):
    
    #Email
    user_input = re.sub(
        r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
        "[HIDDEN]",
        user_input
    )

     # Phone Number
    user_input = re.sub(
        r"\b\d{10}\b",
        "[HIDDEN]",
        user_input
    )

    return user_input
```

```python
text = """
My email is prashasthiboin@gmail.com
Phone : 9874848745
"""

print(pii_middleware(text))
```
